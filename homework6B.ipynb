```stata
clear
```

```stata
set obs 5
```

```stata
global rho = 2/3
global r
global utility = 3
```

```stata
// Risk
gen r = 0 in 1
replace r = 0.25 in 2
replace r = 0.5 in 3
replace r = 0.75 in 4
replace r = 1 in 5
```

```stata
gen wage_u3 = ($utility / $rho ^r)^2 
```

```stata
global utility = 4
gen wage_u4 = ($utility / $rho ^r)^2
```

```stata
twoway line wage_u3 r || line wage_u4 r
```

The equilibirum wage-risk relationship is:
$ W = \Big[ \frac{9}{(\frac{2}{3})^{2r}} \Big]$

```stata
global rho = .2
gen wage = ($utility / $rho ^r)^2 
global rho = .3
gen wage2 = ($utility / $rho ^r)^2 
global rho = .4
gen wage3 = ($utility / $rho ^r)^2 
global rho = .5
gen wage4 = ($utility / $rho ^r)^2 
global rho = .6
gen wage5 = ($utility / $rho ^r)^2 
```

```stata
twoway line wage r || line wage2 r || line wage3 r || line wage4 r || line wage5 r
```

As $ \rho $ increase, the wages decrease. The wage-risk relationship becomes less steep. As the discount factor increase, the compensating wage differentials for dangerous work decreases.
